```python
import pandas as pd 
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
movies=pd.read_csv('my_movies.csv')
movies
```

### PRE - PROCESSING

```python
df=pd.get_dummies(movies)
df
```

### APRIORI ALGORITHM

#### Rules with 20% support

```python
frequent_itemsets1=apriori(df=df,min_support=0.2,use_colnames=True)
pd.set_option('display.max_rows',None)
frequent_itemsets1
```

```python
frequent_itemsets2=apriori(df=df,min_support=0.4,use_colnames=True)
pd.set_option('display.max_rows',None)
frequent_itemsets2
```
