### Test for NORMALITY OF RESIDUALS

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q')
qqplot
#qq plot gives info about the outliers in the data

#qqplot stands for quantile quantile plot
#the model that we have developed, the residuals are not following the normal distribution it their mean value is not zero and dont have a constant variance from mean value 
```

```python
list(np.where(model.resid>50)) 
```

### Residual Plot for homoscedasticity

```python
#its a model validaion technique , remember- lesser the patterns,lesser the problems
#we are defining a standardizing function to standardize the values for plotting purpose
def standardize_val(x):
    return((x-x.mean())/x.std())
```

```python
plt.scatter(standardize_val(model.fittedvalues),standardize_val(model.resid))
plt.title("Residual Plot")
plt.xlabel("Fitted Values")
plt.ylabel("Residual Values")
plt.show()
#we observe that most observations have a mean value=0; and most of them have 
#constant variance from mean
#since the patterns are less, lesser are the problems
```
